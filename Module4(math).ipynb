```python
#1
data = [3,2,5,4,7,2,3,3,1,6,4,2,3,5,2,4,2,1,3,5,6,3,2,1,4,2,4,5,3,2,7,2,3,4,5,1,6,2,4,3,5,3,2,4,2,6,3,2,4,5]

import statistics as stats
from collections import Counter


mean_val = sum(data) / len(data)


median_val = stats.median(data)

freq = Counter(data)
mode_val = [k for k, v in freq.items() if v == max(freq.values())]

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)

print('Average rental ≈ 3.44 days',
       '\nTypical rental = 3 days',
       '\nMost common rental = 2 days')
```

```python
#2
delivery_times = [
    3, 5, 2, 4, 6, 2, 3, 4, 2, 5,
    7, 2, 3, 4, 2, 4, 2, 3, 5, 6,
    3, 2, 1, 4, 2, 4, 5, 3, 2, 7,
    2, 3, 4, 5, 1, 6, 2, 4, 3, 5,
    3, 2, 4, 2, 6, 3, 2, 4, 5, 3
]

range_val = max(delivery_times) - min(delivery_times)

'''
variance formula
mean = sum(delivery_times) / len(delivery_times)
variance = sum((x - mean) ** 2 for x in delivery_times) / len(delivery_times)
'''

import statistics

variance = statistics.pvariance(delivery_times)
std_dev = statistics.pstdev(delivery_times)

print("Range:", range_val)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
#12
# H0: μ=500
# H1: μ!=500
import math

mean = 510
mu = 500
s = 20
n = 25

t = (mean - mu) / (s / math.sqrt(n))

print("t-critical value: 2.064")
print("t-value:", t)
print("\nThe calculated t-value (2.5) is greater than the critical value (2.064), so we reject the null hypothesis. This indicates that there is sufficient evidence to conclude that the average weight of the product is significantly different from 500 grams. Therefore, the company’s claim is not supported by the sample data.")
```

```python
#13
# H0: mu1 = mu2  (no difference between new and traditional method)
# H1: mu1 != mu2 (there is a significant difference)

from scipy.stats import ttest_ind

new_method = [78, 82, 85, 88, 90, 92, 95, 89, 87, 91,
              84, 86, 88, 90, 93, 94, 96, 97, 89, 88,
              85, 87, 90, 91, 92]

traditional = [70, 72, 75, 78, 80, 82, 79, 77, 76, 81,
               74, 73, 75, 78, 79, 80, 82, 83, 77, 76,
               75, 78, 79, 80, 81]

t_stat, p_value = ttest_ind(new_method, traditional)

print("t-statistic:", t_stat)
print("p-value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Reject H0: There is a significant difference between the two methods")
else:
    print("Fail to reject H0: No significant difference")
```

```python
#14
import math

n = 500
x = 320
p_hat = x / n
z = 1.645

SE = math.sqrt((p_hat * (1 - p_hat)) / n)

ME = z * SE

lower = p_hat - ME
upper = p_hat + ME

print("Confidence Interval:", (lower, upper))
```
